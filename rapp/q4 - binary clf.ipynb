## Correlation between features and target

Find features with high correlation with target.

### Consider target as a numeric variable

Then we can use standard correlation coef.

```python
RED_BLUE = plt.get_cmap('RdBu')
def plot_corr_matrix(data, annotate=False, color_map=RED_BLUE, figsize=(10, 10)):
    sns.set_style('white')
    # Compute the correlation matrix
    corr = data.corr()

    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr, dtype=np.bool))

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=figsize)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask,
                cmap=color_map, cbar_kws={"shrink": .5},
#                 vmax=.3, 
                center=0,
                square=True, linewidths=.5,
                annot=annotate
                )
```

```python
plot_corr_matrix(data, annotate=True, figsize=(20, 20))
```

# Predictive models

## Logistic regressoin - base model

I pick logistic regression as this is a binary classification.

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import plot_roc_curve
```

```python
cv = StratifiedKFold(n_splits=5)

def cal_roc_auc(clf, X, y):
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)

    fig, ax = plt.subplots()
    for i, (train_idx, valid_idx) in enumerate(cv.split(X, y)):
        clf.fit(X.iloc[train_idx], y[train_idx])

        viz = plot_roc_curve(clf, X.iloc[valid_idx], y[valid_idx],
                             name='ROC fold {}'.format(i),
                             alpha=0.3, lw=1, ax=ax)
        interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
        interp_tpr[0] = 0.0
        tprs.append(interp_tpr)
        aucs.append(viz.roc_auc)

    return aucs, fig
```

```python
log_reg = LogisticRegression()
log_reg.fit(X, y)
log_reg.score(X, y)
```

```python
aucs, fig = cal_roc_auc(log_reg, X, y)
print('avg AUC:', np.mean(aucs))
```

### Tune hyperparam of RF

```python
from pprint import pprint

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = [1., 'auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(2, 20, num = 10)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
random_rf = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, 
                               n_iter = 100, cv = 3, verbose=2, random_state=42, 
                               n_jobs = -1)


random_rf.fit(X, y)

pprint(random_rf.best_params_)

random_best = random_rf.best_estimator_
random_best.score(X, y)
```

```python
cal_roc_auc(random_best, X, y)
```
